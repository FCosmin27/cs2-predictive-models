```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/cleaned_rounds_data.csv')
df.head()
```

```python
def get_unique_teams(row):
    teams = {row[f'player_{i}_team_name'] for i in range(1, 11) if f'player_{i}_team_name' in row}
    return list(teams)

df['unique_teams'] = df.apply(get_unique_teams, axis=1)
```

```python
df.head()
```

```python
with open('../data/players_stats.json') as file:
    players_stats = json.load(file)

with open('../data/teams_stats.json') as file:
    teams_stats = json.load(file)

teams_stats = {k.lower(): v for k, v in teams_stats.items()}
players_stats = {k.lower(): v for k, v in players_stats.items()}
```

```python
def team_stats_lookup(teams, stat):
    return [teams_stats.get(team.lower(), {}).get(stat, 'NA') for team in teams]
```

```python
df['team_1_position'] = df['unique_teams'].apply(lambda teams: team_stats_lookup(teams, 'position')[0] if len(teams) > 0 else 'NA')
df['team_1_win_percentage'] = df['unique_teams'].apply(lambda teams: team_stats_lookup(teams, 'win_percentage')[0] if len(teams) > 0 else 'NA')
df['team_2_position'] = df['unique_teams'].apply(lambda teams: team_stats_lookup(teams, 'position')[1] if len(teams) > 0 else 'NA')
df['team_2_win_percentage'] = df['unique_teams'].apply(lambda teams: team_stats_lookup(teams, 'win_percentage')[1] if len(teams) > 0 else 'NA')
```

```python
df.head()
```

```python
insert_pos = df.columns.get_loc('round_winner') + 1

team_stats_cols = ['team_1_position', 'team_1_win_percentage', 'team_2_position', 'team_2_win_percentage']

for col in reversed(team_stats_cols):
    df.insert(insert_pos, col, df.pop(col))
```

```python
df.head()
```

```python
df.drop(columns=['unique_teams'], inplace=True)
```

```python
df.head()
```

```python
all_new_columns = {}

for i in range(1, 11):
    if f'player_{i}_name' in df.columns:
        new_columns_df = pd.DataFrame(index=df.index)
        
        new_columns_df[f'player_{i}_rating'] = df[f'player_{i}_name'].str.lower().map(lambda x: players_stats.get(x, {}).get('Rating', 'NA'))
        new_columns_df[f'player_{i}_kills_per_round'] = df[f'player_{i}_name'].str.lower().map(lambda x: players_stats.get(x, {}).get('Kills Per Round', 'NA'))
        new_columns_df[f'player_{i}_headshots'] = df[f'player_{i}_name'].str.lower().map(lambda x: players_stats.get(x, {}).get('Headshots', 'NA'))
        new_columns_df[f'player_{i}_deaths_per_round'] = df[f'player_{i}_name'].str.lower().map(lambda x: players_stats.get(x, {}).get('Deaths Per Round', 'NA'))
        new_columns_df[f'player_{i}_rounds_contributed'] = df[f'player_{i}_name'].str.lower().map(lambda x: players_stats.get(x, {}).get('Rounds Contributed', 'NA'))
        
        all_new_columns[f'player_{i}_name'] = new_columns_df
```

```python
for key, value in all_new_columns.items():
    insert_pos = df.columns.get_loc(key) + 1
    df = pd.concat([df.iloc[:, :insert_pos], value, df.iloc[:, insert_pos:]], axis=1)
```
